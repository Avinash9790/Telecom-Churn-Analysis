```python
# Reload fresh data
churn = pd.read_csv("/content/Untitled spreadsheet - Telco Cuustomer Churn.csv")

# Fix Total Charges
churn['Total Charges'] = pd.to_numeric(churn['Total Charges'], errors='coerce').fillna(0)

# Fix Churn - it's a string 'Yes'/'No' in raw CSV
churn['Churn'] = churn['Churn'].str.strip().map({'Yes': 1, 'No': 0})

# Verify before running EDA
print(churn['Churn'].unique())        # Should show [0, 1]
print(churn['Total Charges'].isnull().sum())  # Should show 0
```

```python
# 1. CHURN DISTRIBUTION
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
churn['Churn'].value_counts().sort_index().plot(kind='bar', ax=axes[0],
                                                 color=['#2ecc71', '#e74c3c'], edgecolor='black')
axes[0].set_title('Churn Distribution')
axes[0].set_xticklabels(['Retained', 'Churned'], rotation=0)
for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width()/2, p.get_height()),
                     ha='center', va='bottom')
churn['Churn'].value_counts().plot.pie(labels=['Retained', 'Churned'],
                                        autopct='%1.1f%%',
                                        colors=['#2ecc71', '#e74c3c'], ax=axes[1])
axes[1].set_title('Churn Percentage')
axes[1].set_ylabel('')
plt.tight_layout()
plt.savefig('1_churn_distribution.png', dpi=150)
plt.show()
```

```python
# ============================================================
# 1. CHURN DISTRIBUTION
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
colors = ['#2ecc71', '#e74c3c']
churn['Churn'].value_counts().sort_index().plot(kind='bar', ax=axes[0], color=colors, edgecolor='black')
axes[0].set_title('Churn Distribution')
axes[0].set_xticklabels(['Retained', 'Churned'], rotation=0)
for p in axes[0].patches:
    axes[0].annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width()/2, p.get_height()),
                     ha='center', va='bottom')
churn['Churn'].value_counts().plot.pie(labels=['Retained', 'Churned'],
                                        autopct='%1.1f%%', colors=colors, ax=axes[1])
axes[1].set_title('Churn Percentage')
axes[1].set_ylabel('')
plt.tight_layout()
plt.savefig('1_churn_distribution.png', dpi=150)
plt.show()

# ============================================================
# 2. DEMOGRAPHIC ANALYSIS
# ============================================================
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
for ax, col in zip(axes, ['Gender', 'SeniorCitizen', 'Partner']):
    churn_rate = churn.groupby(col)['Churn'].mean() * 100
    churn_rate.plot(kind='bar', ax=ax, color='#3498db', edgecolor='black')
    ax.set_title(f'Churn Rate by {col}')
    ax.set_ylabel('Churn Rate (%)')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width()/2, p.get_height()),
                    ha='center', va='bottom')
plt.tight_layout()
plt.savefig('2_demographic_churn.png', dpi=150)
plt.show()

# ============================================================
# 3. CONTRACT & PAYMENT ANALYSIS
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for ax, col in zip(axes, ['Contract', 'Payment Method']):
    churn_rate = churn.groupby(col)['Churn'].mean() * 100
    churn_rate.sort_values(ascending=False).plot(kind='bar', ax=ax,
                                                  color='#e74c3c', edgecolor='black')
    ax.set_title(f'Churn Rate by {col}')
    ax.set_ylabel('Churn Rate (%)')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=15, ha='right')
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width()/2, p.get_height()),
                    ha='center', va='bottom')
plt.tight_layout()
plt.savefig('3_contract_payment_churn.png', dpi=150)
plt.show()

# ============================================================
# 4. TENURE ANALYSIS
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for val, color, label in zip([0, 1], ['#2ecc71', '#e74c3c'], ['Retained', 'Churned']):
    subset = churn[churn['Churn'] == val]
    axes[0].hist(subset['tenure'], bins=30, alpha=0.6, color=color, label=label, edgecolor='black')
axes[0].set_title('Tenure Distribution by Churn')
axes[0].set_xlabel('Tenure (Months)')
axes[0].legend()

churn['tenure_group'] = pd.cut(churn['tenure'], bins=[0, 12, 24, 36, 48, 60, 72],
                                labels=['0-12', '12-24', '24-36', '36-48', '48-60', '60-72'])
tenure_churn = churn.groupby('tenure_group', observed=True)['Churn'].mean() * 100
tenure_churn.plot(kind='bar', ax=axes[1], color='#9b59b6', edgecolor='black')
axes[1].set_title('Churn Rate by Tenure Group')
axes[1].set_ylabel('Churn Rate (%)')
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=0)
for p in axes[1].patches:
    axes[1].annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width()/2, p.get_height()),
                     ha='center', va='bottom')
plt.tight_layout()
plt.savefig('4_tenure_churn.png', dpi=150)
plt.show()

# ============================================================
# 5. CHARGES ANALYSIS
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
for ax, col in zip(axes, ['Monthly Charges', 'Total Charges']):
    for val, color, label in zip([0, 1], ['#2ecc71', '#e74c3c'], ['Retained', 'Churned']):
        subset = churn[churn['Churn'] == val]
        ax.boxplot(subset[col], positions=[val], patch_artist=True,
                   boxprops=dict(facecolor=color, color='black'))
    ax.set_title(f'{col} by Churn')
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['Retained', 'Churned'])
plt.tight_layout()
plt.savefig('5_charges_churn.png', dpi=150)
plt.show()

# ============================================================
# 6. SERVICES ANALYSIS
# ============================================================
services = ['Phone Service', 'Multiple Lines', 'Internet Service',
            'Online Security', 'Online Backup', 'Device Protection',
            'Tech Support', 'Streaming TV', 'Streaming Movies']
fig, axes = plt.subplots(3, 3, figsize=(18, 14))
axes = axes.flatten()
for ax, col in zip(axes, services):
    churn_rate = churn.groupby(col)['Churn'].mean() * 100
    churn_rate.plot(kind='bar', ax=ax, color='#e67e22', edgecolor='black')
    ax.set_title(f'Churn Rate by {col}')
    ax.set_ylabel('Churn Rate (%)')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=15, ha='right')
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width()/2, p.get_height()),
                    ha='center', va='bottom')
plt.tight_layout()
plt.savefig('6_services_churn.png', dpi=150)
plt.show()

# ============================================================
# 7. PAPERLESS BILLING & DEPENDENTS
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for ax, col in zip(axes, ['Paperless Billing', 'Dependents']):
    churn_rate = churn.groupby(col)['Churn'].mean() * 100
    churn_rate.plot(kind='bar', ax=ax, color='#3498db', edgecolor='black')
    ax.set_title(f'Churn Rate by {col}')
    ax.set_ylabel('Churn Rate (%)')
    ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.1f}%', (p.get_x() + p.get_width()/2, p.get_height()),
                    ha='center', va='bottom')
plt.tight_layout()
plt.savefig('7_paperless_dependents_churn.png', dpi=150)
plt.show()

# ============================================================
# 8. CORRELATION HEATMAP
# ============================================================
num_cols = ['tenure', 'Monthly Charges', 'Total Charges', 'Churn', 'SeniorCitizen']
plt.figure(figsize=(8, 6))
sns.heatmap(churn[num_cols].corr(), annot=True, fmt='.2f',
            cmap='RdYlGn', linewidths=0.5, square=True)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('8_correlation_heatmap.png', dpi=150)
plt.show()

# ============================================================
# 9. SUMMARY
# ============================================================
print(f"Total Customers       : {len(churn)}")
print(f"Churned               : {int(churn['Churn'].sum())}")
print(f"Retained              : {int((churn['Churn'] == 0).sum())}")
print(f"Churn Rate            : {churn['Churn'].mean()*100:.2f}%")
print(f"Avg Monthly (Churned) : ${churn[churn['Churn']==1]['Monthly Charges'].mean():.2f}")
print(f"Avg Monthly (Retained): ${churn[churn['Churn']==0]['Monthly Charges'].mean():.2f}")
print(f"Avg Tenure (Churned)  : {churn[churn['Churn']==1]['tenure'].mean():.1f} months")
print(f"Avg Tenure (Retained) : {churn[churn['Churn']==0]['tenure'].mean():.1f} months")
```
